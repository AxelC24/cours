```python
import pandas as pd

url = "https://docs.google.com/spreadsheets/d/1Mr-gVp_I2VZCVHyb8Oo7U4dr4o9zMCLXOXrnPrTgeTg/export?format=csv&gid=810928449"
df = pd.read_csv(url)
```

```python
# On peut utiliser chat gpt avec le prompt suivant :
# https://chat.openai.com/share/b017492b-51bc-475f-9cda-e3c6b5231805
df.rename(columns={    "Timestamp": "timestamp",
    "email": "email",
    "Prénom": "first_name",
    " Nom": "last_name",
    "Github username": "github",
    "Votre niveau en programmation/Python": "python_skill_level",
    "Selon vous, dans votre vie professionnelle future, vous utiliserez la programmation ?": "will_use_programming",
    "Qu'est ce que je souhaite apprendre/comprendre/découvrir avec ce cours ?": "learning_objectives",
    "J'ai déja entendu parlé de :": "heard_of",
    "Je mesure (en cm)": "height_cm",
    "Mon poids (en kg)": "weight_kg",
    "Je suis ": "gender",
    "Ma matière préférée est :": "favorite_subject",
    "Ma date de naissance (jj/mm/aaaa)": "birthdate",
    "Des idées de sujet pour analyse de donnée ?": "data_analysis_ideas"
}, inplace=True)
```
